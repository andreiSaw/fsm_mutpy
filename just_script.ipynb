```python
data_in = []
data_out = []
with open("0.seq",'r') as input:
    for line in input:
        l1 = line.split(" ")
        l1 = [d.replace("A", "Value.A") for d in l1]
        l1 = [d.replace("B", "Value.B") for d in l1]
        l2 = []
        l3 = []
        for v in l1:
            l2.append(int(v.split('/')[1][0]))
            l3.append(v[:8])
        data_in.append(l3)
        data_out.append(l2)
```
